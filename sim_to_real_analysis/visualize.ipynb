```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("joint1_pos_data.csv")
df.head(), len(df)
```

```python
times = [0.0]*len(df)
for i in range(1, len(df)):
    times[i] = round(times[i-1] + 0.04, 2) 
times
```

```python
import matplotlib.pyplot as plt
plt.plot(times, df["joint1_pos"], label="Joint 1 Position")
plt.xlabel("Time (s)")
plt.ylabel("Position (rad)")
plt.title("Joint 1 Position Over Time")
```
