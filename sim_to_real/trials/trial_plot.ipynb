```python
import pandas as pd
```

```python
df = pd.read_csv("trials.csv")
df
```

```python
import matplotlib.pyplot as plt

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(df['labels'], df['success_rate (%)'], color='skyblue')
plt.title('Success Rate by Labels')
plt.xlabel('Labels')
plt.ylabel('Success Rate (%)')

# Save the plot to a file
plt.savefig('success_rate_plot.png')

# Show the plot
plt.show()
```
